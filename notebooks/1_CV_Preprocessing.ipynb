# Image Preprocessing for the training phase

With the goal of minimize the processing power, and time, needed during the training phase, we’ll make some modifications to the observations of the “Env” object.

The modifications that will be implemented will be mainly visual:


* Transform to grayscale the frames.
* Create a __focus area__, that reduces the visual area to mainly where the action takes place.

```python
# Importaciones
import retro
import cv2
import matplotlib.pyplot as plt
import numpy as np

STATE = 'KingHippo.state'
env = retro.make(game='Mtpo-Nes'
                 , state=STATE
                )
```

First, lets get an idea of the screen:

```python
env.reset()
env_screen = env.render(mode ='rgb_array')
env.close()
plt.imshow(env_screen)
plt.axis('off')
plt.show()
```

```python
env_screen.shape
```

We can clearly see that the original screen has a size as previously mentioned of 224x240x3.
Much of the space of the screen doesn´t contain useful information, we could safely say than 2/3 of the screen are useless in terms of the visual information necessary for the player. So we can safely remove that part to save space, and leave a “focus area” as mentioned before, where only Little Mac and the enemy are visible:

```python
# Separeates the screen dimensions
xlen = env_screen.shape[0]
ylen = env_screen.shape[1]

# Cuts the top 20 % space, and 2/3 of the sides of the screen
cropped_env_screen = env_screen[int(xlen*(1/8)):int(xlen*(3/2)),int(ylen/3):-int(ylen/3)]
plt.imshow(cropped_env_screen)
plt.axis('off')
plt.show()
```

Now, we reduce the color channel quantity to just one (1), this is also known as changing the color to "gray scales":

```python
# Turns to "gray scale"
gray_cropped_env_screen = cv2.cvtColor(cropped_env_screen, cv2.COLOR_BGR2GRAY)
gray_cropped_env_screen = (gray_cropped_env_screen - np.min(gray_cropped_env_screen)) / (np.max(gray_cropped_env_screen) - np.min(gray_cropped_env_screen))

# Shows the image
plt.imshow(gray_cropped_env_screen, cmap='gray')
plt.axis('off')
plt.show()
```

```python
gray_cropped_env_screen.shape
```

As a final step, we'll reduce the previous image to a 84x84x1 dimensions image, as the algorithms we'll implement favor "square" input images.

```python
resize = cv2.resize(gray_cropped_env_screen, (84,84), interpolation=cv2.INTER_CUBIC)
plt.imshow(resize, cmap='gray')
plt.axis('off')
plt.show()
```
